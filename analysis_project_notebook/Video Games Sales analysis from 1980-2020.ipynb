## Data Preparation and Cleaning

**TODO** - Write some explanation here.



````python
import numpy as np
import pandas as pd
````

````python
videogame_df = pd.read_csv('videogamesales/vgsales.csv')
videogame_df.head(10)
````

````python
print(videogame_df.shape)
print(videogame_df.size)
````

````python
videogame_df.info()
````

````python
videogame_df.describe()
````

````python
videogame_df.columns
````

````python
videogame_df.isna
````

````python
# This are all the Df having value as Nan.
videogame_df[videogame_df.isna().any(axis=1)]
````

````python
# This are the Df where Publisher value is Nan
videogame_df[videogame_df['Publisher'].isna()]
````

````python
# This are the Df where Year value is Nan
videogame_df[videogame_df['Year'].isna()]
````

````python
# This is the df where the Publisher is Unknown
unknown_publishers_df = videogame_df[videogame_df['Publisher'] == 'Unknown']
````

````python
# Making the Unknown and Nan values as Anonymous
videogame_df = videogame_df.replace(['Unknown','Nan'], 'Anonymous')
````

````python
# Now the value of sale of video games accross the globe and countries is in millions. Hence, convert them into actual value rather than in decimal form
videogame_df[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']]*=1000000
videogame_df
````

````python
videogame_df.sort_values(by='Year').tail(250)
````

````python
year_df = videogame_df.groupby('Year').count()
year_df
````

````python
# Here interpolate method is used that fills the Nan values based on the previous or next value linearly.
# limit_direction means considering previous value and move forward.
videogame_df['Year'] = videogame_df['Year'].interpolate(method ='linear', limit_direction ='forward')
videogame_df.tail(20)
````

````python
# As you can see that some of the year values are not in the required format also they are in type float.
videogame_df['Year'].unique().shape
````

Now the is relevant and ready to perform analysis and visualtions.

````python
# Hence we will convert type of Year to Integer.
videogame_df['Year'] = videogame_df['Year'].astype(int)
````

````python
videogame_df.sort_values(by='Year')
````

## Exploratory Analysis and Visualization

**TODO** - Using the cleaned data, now we can visualize the different aspects of the data and compare them with other values in the Dataframe.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

**TODO** - Defining the variables to plot graphs

````python
years = videogame_df['Year']
na_sale =  videogame_df['NA_Sales'].astype(int)
print(na_sale)
eu_sale =  videogame_df['EU_Sales'].astype(int)
jp_sale =  videogame_df['JP_Sales'].astype(int)
global_sale =  videogame_df['Global_Sales'].astype(int)
````

````python
plt.bar(years, na_sale)
````

Here we can observe that After year 2005 i.e in year 2006, thats the highest number of sale for a game. Between year 1991-2004 the was unpredictable and has not been very good one. From 2015-2020, the sale is almost lowest.

````python
plt.plot(years,global_sale)
````

**TODO** - From the above figure we can see that the sale between 1995-2007 has large number of sale and rest the sale lies between 2-4 e7.

````python
sns.scatterplot(x=years,y=na_sale)
sns.scatterplot(x=years,y=eu_sale)
sns.scatterplot(x=years,y=jp_sale)
plt.legend(labels=["NA-Sale","Eu-Sale","JP-Sale"])
````

As one can see, Sales in NA is greater as compared to other two.
Sale in JP is lowest compare to other two where as EU lies in between of both.

## Asking and Answering Questions

TODO - Since the data has been clean and we also visualized it, now we need to ask some question and get some inference from this data.



#### Q1: TODO - Which are the highest selling top 3 games around the globe?

````python
highest_game_df = videogame_df.sort_values(by='Global_Sales',ascending=False).head(3)
highest_game_df
````

#### Q2: TODO - In which year, the sales across the globe is smallest and for which game?

````python
# highest_sale_year = videogame_df.groupby('Year')['Global_Sales'].sum()
# highest_sale_year = videogame_df.sort_values('Global_Sales',ascending=False).groupby('Year').head(3)
highest_sale_year = videogame_df.groupby(["Year","Name"])["Global_Sales"].sum().nsmallest(1)     
highest_sale_year
````

#### Q3: TODO - Which games has higher sales in NA compare to JP and EU combined?

````python
na_sale_df = videogame_df[(videogame_df['NA_Sales']) > (videogame_df['JP_Sales'] + videogame_df['EU_Sales'])]['Name']
na_sale_df
````

#### Q4: TODO - Which are the top 3 Platforms that are successful and to have largest global sales?

````python
 videogame_df.groupby(["Platform"])["Global_Sales"].sum().nlargest(3)  
````

#### Q5: TODO - Which game has the Lowest selling percentage of game around the globe compare to the highest selling games around the globe?

````python
videogame_df.head()
````

````python
# Creating a new column which defines the percentage of sale compare to the highest selling game around globe
videogame_df['Sale_Ratio in %'] = (videogame_df['Global_Sales'] / videogame_df.Global_Sales.max()) * 100 
````

````python
videogame_df.sort_values(by='Sale_Ratio in %',ascending=True).head(1)
````
